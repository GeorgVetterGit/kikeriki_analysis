Ausstattung: 5 Spielfiguren, 18 Tierkarten, 1 Farbwürfel
Aufmachung: Die Spielfiguren sind groß und in fünf Farben lackiert. Diese Farben gibt es auch auf dem Würfel zu sehen, der zusätzlich noch ein lachendes Gesicht zeigt. Die Tierkarten zeigen verschiedene Tiere. Jedes Tier ist zweimal vorhanden. Neben der Abbildung findet man noch drei verschiedenfarbige Punkte unter den Motiven. Die Punkte entsprechen in den Farben den Spielfiguren. Das gesamte Material ist in einer kleinen Schachtel untergebracht, die gut gefüllt ist.
Ziel: Es gibt vier verschiedene Spielmöglichkeiten.
Das erste Spiel nennt sich „Kikeriki". Die Spielfiguren und der Würfel kommen dabei zunächst in die Spielmitte. Dann nimmt man sechs beliebige Tierkarten und legt diese als verdeckten Stapel daneben. Zum Abschluß der Vorbereitungen wird schließlich die oberste Tierkarte aufgedeckt und man einigt sich auf die Reihenfolge im Spiel.
Der aktive Spieler darf in seinem Zug einmal würfeln. Zeigt dieser eine Farbe, die auch auf dem Kärtchen als Farbpunkt zu sehen ist, wird die entsprechende Spielfigur dort abgestellt. Anschließend darf das Kind erneut würfeln. Ist die Farbe zwar auf der Karte zu sehen, aber die Spielfigur steht bereits auf dem Feld, ist der folgende Spieler am Zug. Dies gilt auch, wenn die gewürfelte Farbe nicht auf dem Kärtchen als Farbpunkt zu erkennen ist.
Das lachende Gesicht ist ein Joker. Das Kind darf eine beliebige passende Figur auf die Karte stellen und ist sofort noch einmal an der Reihe.
Wenn alle drei passenden Figuren auf einer Karte stehen, erhält der aktive Spieler diese Karte und legt sie vor sich ab. Mit dem Nehmen der Karte endet der Spielzug automatisch.
Spielende: Sobald alle sechs Kärtchen verteilt sind, endet das Spiel. Es gewinnt der Spieler mit den meisten Tierkarten. Bei Gleichstand gewinnen alle diese Spieler.

Import Libraries

```python
import random
import scipy.stats as stats
```

Set Parameters

```python
DICE = ['red', 'green', 'blue', 'yellow', 'purple', 'smiley']
NO_PLAYERS = 2
N_CARDS = 6
COLORS = ['red', 'green', 'blue', 'yellow', 'purple']

ITERATIONS = 10000
```

Helper functions

```python
def roll_dice():
    return random.choice(DICE)

def create_card_deck(N_CARDS: int = 6):
    deck = []
    for _ in range(N_CARDS):
        random.shuffle(COLORS)
        deck.append(COLORS[:3])
    return deck
```

Beispielspielablauf

```python
deck = create_card_deck()
player = random.randint(0, NO_PLAYERS - 1)
scores = [0] * NO_PLAYERS

print("Starting the game with", NO_PLAYERS, "players and", N_CARDS, "cards.")
print("Deck:", deck)
print(f"Player {player + 1}'s turn.")

while len(deck) > 0:
    throw = roll_dice()
    print("Player ", player + 1, " rolled ", throw)
    if throw == 'smiley':
        throw = deck[0][0]
    if throw in deck[0]:
        print(f"Player {player + 1} hits on {throw}!")
        deck[0].remove(throw)
        print("Remaining on card:", deck[0])
    else:
        print(f"Player {player + 1} misses.")
        if player + 1 < NO_PLAYERS:
            player += 1
        else:
            player = 0
    if len(deck[0]) == 0:
        print(f"Player {player + 1} completed a card!")
        deck.pop(0)
        scores[player] += 1

print("Game over!")
print("Final scores:", scores)
```

Simulation function

```python
def simulate_game(change_player_after_card: bool = False, NO_PLAYERS: int = 2, N_CARDS: int = 6)->list:

    scores_overall = []

    for _ in range(ITERATIONS):

        deck = create_card_deck(N_CARDS)
        player = 0 #random.randint(0, NO_PLAYERS - 1)
        scores = [0] * NO_PLAYERS

        while len(deck) > 0:
            throw = roll_dice()
            if throw == 'smiley':
                throw = deck[0][0]
            if throw in deck[0]:
                deck[0].remove(throw)
            else:
                if player + 1 < NO_PLAYERS:
                    player += 1
                else:
                    player = 0
            if len(deck[0]) == 0:
                deck.pop(0)
                scores[player] += 1
                if change_player_after_card:
                    if player + 1 < NO_PLAYERS:
                        player += 1
                    else:
                        player = 0

        scores_overall.append(scores)

    return scores_overall
```

```python
scores_no_change = simulate_game(change_player_after_card=False)
scores_change = simulate_game(change_player_after_card=True)

player1_no_change = [score[0] for score in scores_no_change]
player1_change = [score[0] for score in scores_change]
player2_no_change = [score[1] for score in scores_no_change]
player2_change = [score[1] for score in scores_change]

# Compare strategies
t_stat, p_value = stats.ttest_ind(player1_no_change, player1_change)
print("T-test results comparing strategies:")
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.01:
    print("The difference between the two strategies is statistically significant.")
else:
    print("No statistically significant difference between the two strategies.")

# Compare players in no change strategy
t_stat_players, p_value_players = stats.ttest_ind(player1_no_change, player2_no_change)
print("\nT-test results comparing Player 1 and Player 2 in no change strategy:")
print("T-statistic:", t_stat_players)
print("P-value:", p_value_players)
if p_value_players < 0.01:
    print("The difference between Player 1 and Player 2 is statistically significant.")
else:
    print("No statistically significant difference between Player 1 and Player 2.")

# Compare players in change strategy
t_stat_players_change, p_value_players_change = stats.ttest_ind(player1_change, player2_change)
print("\nT-test results comparing Player 1 and Player 2 in change strategy:")
print("T-statistic:", t_stat_players_change)
print("P-value:", p_value_players_change)
if p_value_players_change < 0.01:
    print("The difference between Player 1 and Player 2 is statistically significant.")
else:
    print("No statistically significant difference between Player 1 and Player 2.")
```

Simulation study

```python
#Parameters
NO_PLAYERS = list(range(2, 6))
N_CARDS = list(range(2, 19))
```

```python
significance_results = []

for n_players in NO_PLAYERS:
    for n_cards in N_CARDS:
        scores_no_change = simulate_game(change_player_after_card=False, NO_PLAYERS=n_players, N_CARDS=n_cards)
        scores_change = simulate_game(change_player_after_card=True, NO_PLAYERS=n_players, N_CARDS=n_cards)

        player1_no_change = [score[0] for score in scores_no_change]
        player1_change = [score[0] for score in scores_change]
        player2_no_change = [score[1] for score in scores_no_change]
        player2_change = [score[1] for score in scores_change]

        #strategy comparison	
        t_stat, p_value = stats.ttest_ind(player1_no_change, player1_change)
        significance_results.append(('strategy',n_players, n_cards, t_stat, p_value))

        #player comparison no change
        t_stat_players, p_value_players = stats.ttest_ind(player1_no_change, player2_no_change)
        significance_results.append(('no_change_player',n_players, n_cards, t_stat_players, p_value_players))

        #player comparison change
        t_stat_players_change, p_value_players_change = stats.ttest_ind(player1_change, player2_change)
        significance_results.append(('change_player',n_players, n_cards, t_stat_players_change, p_value_players_change))
        
import pandas as pd
df_results = pd.DataFrame(significance_results, columns=['comparison', 'n_players', 'n_cards', 't_statistic', 'p_value'])
print("\nSummary of significance tests across different parameters:")
print(df_results[df_results['p_value'] < 0.01])
print(len(df_results[df_results['p_value'] < 0.01]), "out of", len(df_results), "tests showed statistically significant differences.")
```
